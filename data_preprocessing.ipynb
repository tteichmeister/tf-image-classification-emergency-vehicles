```python
download_configs = [
    {
        "keywords":"Feuerwehr Fahrzeug, Feuerwehr Auto",
        "limit": 100,
        "print_urls": True, 
        "output_directory":"images",
        "image_directory":"fire"
    },
    {
        "keywords":"Polizei Fahrzeug, Polizei Auto, Polizeiwagen",
        "limit": 100,
        "print_urls": True, 
        "output_directory":"images",
        "image_directory":"police"
    },
    {
        "keywords":"Rettungs Auto, Rettungswagen, Rettungs Fahrzeug",
        "limit": 100,
        "print_urls": True, 
        "output_directory":"images",
        "image_directory":"rescue"
    }
]

resp = google_images_download.googleimagesdownload()
for cfg in download_configs:
    print(cfg)
    paths  = resp.download(cfg)
    print(paths)
```
